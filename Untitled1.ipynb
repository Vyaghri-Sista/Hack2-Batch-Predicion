#Standardize features by removing the mean and scaling to unit variance

The standard score of a sample x is calculated as:

    z = (x - u) / s


```python
from sklearn.preprocessing import StandardScaler

df1_trans = StandardScaler().fit_transform(df1)
df1_trans.shape
df2=pd.DataFrame(df1_trans,index=[i for i in range(df1_trans.shape[0])],columns=['x_'+str(i) for i in range(df1_trans.shape[1]) ])
```

```python
# sns.distplot(df1_trans['y2'],bins=30)
df2.head()
```

```python
df3=df2.iloc[:,0:-1]
df3.shape
```

```python
data = np.array(df3)
data=np.transpose(data)
covMatrix = np.cov(data,bias=True)
covMatrix.shape
print (covMatrix)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
principalComponents = pca.fit_transform(data)
```

```python
u, s, vh = np.linalg.svd(covMatrix, full_matrices=True)
```
